```python
import sympy as sp
from sympy import symbols
from numpy import linspace
from sympy import lambdify
from sympy import I as j
from sympy import re, im
import matplotlib.pyplot as mpl


from sympy.physics.mechanics import *
init_vprinting()
```

## 1.D-2

```python
rho_0 = symbols('rho_0')
c_0 = symbols('c_0')
r_0 = symbols('r_0')
k = symbols('k')
```

```python
Z = rho_0*c_0 / (1 + ( 1 / (j*k*r_0)) )
Z
```

```python
res = {c_0: 1, rho_0: 1, r_0: 1}

plot1 = sp.plot(re(Z.subs(res)), (k, 0, 10))
plot2 = sp.plot(im(Z.subs(res)), (k, 0, 10))
plot1.extend(plot2)
plot1.show()
```

```python
# Imaginary part check
Z2 = c_0*k*rho_0*r_0 / (k**2 * r_0**2 + 1)
sp.plot(Z2.subs(res), (k, 0, 10))
```

```python
# Real part check
Z3 = rho_0*c_0*((k*c_0) / (k*c_0 + 1))
sp.plot(Z3.subs(res), (k, 0, 10))
```
